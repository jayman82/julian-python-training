# ðŸŽ‰ Challenge Clicker (Tkinter)
Click the button to get a challenge!

```
import tkinter as tk

# Function to update the label when the button is clicked
def challenge():
    count.set(count.get() + 1)
    challenge_label.config(text=f"Go Julian! ðŸŽ‰ Clicks: {count.get()}")

# Main window
root = tk.Tk()
root.title("Julian's Challenge Clicker")

count = tk.IntVar(value=0)

challenge_label = tk.Label(root, text="Click to challenge!", font=('Arial', 16))
challenge_label.pack(pady=10)

challenge_button = tk.Button(root, text="Challenge!", command=challenge, font=('Arial', 14), bg='pink')
challenge_button.pack(pady=10)

root.mainloop()
```
